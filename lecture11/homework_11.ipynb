## Problem 1 - Calculating expectations for discrete random variables

Consider the Categorical random variable $X \sim \text{Categorical}(0.2, 0.4, 0.4)$ taking three discrete values $1, 2,$ and $3$.
Find the numerical answer for the following.
Hint: You can do it by hand or write some code. It is up to you.

+ $\mathbf{E}[X] = ?$

**Answer: 2.2 (see code below)**
<br><br><br><br><br><br><br><br>

````python
xs = np.array([1, 2, 3])
ps = np.array([0.2, 0.4, 0.4])
E_X = np.sum(xs * ps)
print(f'E[X] = {E_X:1.2f}')
````

+ $\mathbf{E}[X^2] = ?$

**Answer: 5.4 (see code below)**
<br><br><br><br><br><br><br><br>

````python
E_X2 = np.sum(xs**2 * ps)
print(f'E[X^2] = {E_X2:1.2f}')
````

+ $\mathbf{V}[X] = ?$

**Answer: 0.56**
<br><br><br><br><br><br><br><br>

````python
V_X = E_X2 - E_X**2
print(f'V[X] = {V_X:1.2f}')
````

+ $\mathbf{E}[e^X] = ?$

**Answer: 11.53**
<br><br><br><br><br><br><br><br>

````python
E_eX = np.sum(np.exp(xs) * ps)
print(f'E[e^X] = {E_eX:1.2f}')
````

## Problem 2 - Calculate the expectation and variance of a continuous random variable

Take an exponential random variable, see [Problem 2 of Homework 10](lecture10:homework:problem2),
with rate parameter $\lambda$:

$$
T \sim \text{Exp}(\lambda).
$$

The PDF of the exponential is:

$$
p(t) = \lambda e^{-\lambda t},
$$

for $t\ge 0$ and zero otherwise.

+ Prove mathematically that $\mathbf{E}[T] = \lambda^{-1}$. Hint: You need to do the integral $\int_0^\infty tp(t)dt$ using [integration by parts](https://en.wikipedia.org/wiki/Integration_by_parts).

**Answer: Since E[T] = integral from 0 to infinity *t*p(t)dt. Substituting the PDF, p(t) becomes λe^-λt dt. Then, integration by parts with u = t dv = λe^-λt gives [ite^-λt] evaluated at 0 and infinity, and + integral from 0 to infinity of e^-λt. Solving this, one term is 0 and the integral from 0 to infinity of e^-λt is 1/λ. So finally, E[T] = 1/λ = λ^-1.**
<br><br><br><br><br><br><br><br>

+ Prove mathematically that $\mathbf{V}[T] = \lambda^{-2}$. Hint: Use integration by parts to find $\mathbf{E}[T^2]=\int_0^\infty t^2 p(t)dt$ and then use one of the properties of the variance.

**Answer: We know that V[T] = E[T^2] - (E[T])^2. As such, E[T^2] is the integral from 0 to infinity of t^2(λe^λt)dt. By integration by parts, u = t^2 and dv = λe^-λt. This yields [-t^2e^-λt] evaluated at 0 to infinity, plus the integral from 0 to infinity of 2te^λt dt. So, we can write E[T^2] = 0 + 2/λ(integral from 0 to infinity of t(λe^-λt)) dt, which is just 2/λ E[T]. Now, squaring this to get E[T^2] gives 2/λ * 1/λ = 2/λ^2. So, the Variance V[T] is then 2/λ^2 - (1/λ)^2 which gives the final answer of 1/λ^2.**
