## Problem 1 - Playing with matrices

Write Python code that constructs the 90-degree rotation matrix in 2D:

$$
A = \begin{bmatrix}
    \cos(\pi/2) &  -\sin(\pi/2) \\
    \sin(\pi/2) & \cos(\pi/2)
\end{bmatrix}
$$

````python
import numpy as np

A = np.array([[np.cos(np.pi/2), -np.sin(np.pi/2)], [np.sin(np.pi/2), np.cos(np.pi/2)]])
````

Verify that the matrix $A$ rotates the vector $x=\begin{bmatrix} 1 \\ 0 \end{bmatrix}$ by 90 degrees counterclockwise. That is, $A x = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$.

````python
x = np.array([[1], [0]])
Ax = A.dot(x)
print(Ax)
````

Create the matrix:

$$
B = A\cdot A
$$

````python
B = A.dot(A)
print(B)
````

Is $B$ a rotation matrix? What is the effect of $B$ on the vector $x$? Hint: To verify that $B$ is a rotation matrix, you need to check that $B^T\cdot B = I$ (i.e., that $B$ is orthogonal) and that the determinant of $B$ is 1. You can use the ``np.linalg.det()`` function to compute the determinant of a matrix.

````python
det = np.linalg.det(B)
identityMatrix = np.eye(2)
identityCheck = B.T.dot(B)

print("determinant is:", det)
print("orthogonal check:", identityCheck)
print("identity matrix is: ", identityMatrix)

print("Since B is verified to be orthogonal and the determinant of B is 1, B is a rotation matrix. The effect that this has on B is that it flips the vector, also ")
print("known as rotating it 180 degrees. This would make it point the other direction.")
````

Apply the matrix $B$ to the vector $x$ and verify that it rotates the vector by 180 degrees counterclockwise.

````python
Bx = B.dot(x)
print(Bx)
````

If you perform 90 degree rotation 4 times, what is the effect of the resulting matrix on the vector $x$? Verify your answer by showing that multiplying $A$ by itself 4 times results in the identity matrix.

````python
R1 = A.dot(A)
R2 = R1.dot(A)
R3 = R2.dot(A)
verification = R3.dot(x)
print(verification)
print("As shown by the calculations here, doing this 90 degree rotation 4 times on x would result in the original x.")
````
